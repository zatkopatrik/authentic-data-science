```python
data = pd.read_csv('Data/avocado.csv')
```

```python
# 10 % missing values
data['region'] = data['region'].sample(frac = 0.9)
```

```python
# 20 % missing values
data['Small Bags'] = data['Small Bags'].sample(frac = 0.8)
```

```python
data.drop('Total Bags', axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.to_csv('avocado_missing.csv', index = False)
```
