# Multivariate Analysis with Seaborn

**Intro**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
```

```python
# Load the data
penguins = sns.load_dataset('penguins')
```

**remove missing values**

```python
# Take a look at first 5 rows
penguins.head()
```

## Encoding third variable

key points

```python
# changing position of islands
ordering = ['Biscoe', 'Dream', 'Torgersen']
```

**add counts on top of bars**
**adjust the position of bars**

```python
# changing order of counts
species = penguins.species.sort_values(ascending = False)
```

```python
sns.countplot(data = penguins, x = 'island', hue = species, order = ordering, palette = 'Set3');
```

```python
sns.boxplot(data = penguins, x = 'island', y = 'body_mass_g', hue = 'sex');
```

## relational plots

```python
sns.relplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', hue = 'species');
```

```python
sns.relplot(data = penguins, x = 'bill_length_mm', y = 'bill_depth_mm', hue = 'species');
```

##  FacetGrid

intro

```python
sns.FacetGrid(penguins)
```

```python
g = sns.FacetGrid(penguins, col = 'species', row = 'sex')
g.map(sns.histplot, 'bill_length_mm', color = 'paleturquoise')
```

```python
# Defining colors and markers of gender
c_m = {'color':['lightcoral','royalblue'],
       'markers':['x','o']}

facetgrid = sns.FacetGrid(penguins, col = 'species', row = 'sex', hue_kws = c_m)
facetgrid.map(sns.scatterplot, 'body_mass_g', 'flipper_length_mm');
```

## PairGrid

```python
palette = ['cornflowerblue','lightgreen','gold']
```

```python
sns.set_palette(sns.color_palette(palette))
```

```python
g_grid = sns.PairGrid(penguins, hue = 'species')
g_grid.map_diag(sns.kdeplot, fill = True)
g_grid.map_offdiag(sns.scatterplot)
g_grid.add_legend()
```
